```python
#importing librabries
import pickle
import pandas as pd
from pathlib import Path
```

```python
with open("Raw Json data/raw_data.pkl", "rb") as f:
    raw_data = pickle.load(f)
```

```python
raw_data
```

```python
df = raw_data["very_active_minutes"].copy()
```

```python
df.head()
```

```python
df["dateTime"] = pd.to_datetime(df["dateTime"], errors = "coerce")
```

```python
df["value"] = pd.to_numeric(df["value"], errors="coerce")
```

```python
df = df.dropna(subset=["dateTime", "value"])
```

```python
df = df.sort_values("dateTime").reset_index(drop=True)
```

```python
df.head()
```

```python
cleaned_data = {}
cleaned_data["very_active_minutes"] = df
```

```python
cleaned_data["very_active_minutes"].to_parquet("C:/Users/bisar/My fitbit dashboard project/Cleaned/very_active_minutes.parquet", index=False)
```

```python
cleaned_data["very_active_minutes"].to_csv("C:/Users/bisar/My fitbit dashboard project/Cleaned/very_active_minutes.csv", index=False)
```
