**Total Active Minutes**

```python
df["total_active_minutes"] = (
    df["lightly_active_minutes"]
    + df["moderately_active_minutes"]
    + df["very_active_minutes"]
)
```

**Active vs Sedentary Ratio**

```python
df["active_sedentary_ratio"] = df["total_active_minutes"] / df["sedentary_minutes"]
```

**combining calories, steps, distance into energy & movement KPIs.**

```python
df[["date", "total_active_minutes", "sedentary_minutes", "active_sedentary_ratio"]].head()
```
